### Step 1: Select Data Frame


```python
# Import relevant libraries
import pandas as pd

# Read Data Frame
data = pd.read_excel("nyc-violation-ticket.xlsx")
npv_df = pd.DataFrame(data)
npv_df

# Display Data Frame Info
npv_df.info()
print()

# Display Data Frame
npv_df
```

### Step 2: Filter Data Frame to Include Specific Columns

```python
# Assign columns_to_keep as a list of columns you want to keep
columns_to_keep = ["Plate ID", "Registration State", "Vehicle Make", 
                   "Vehicle Color", "Violation Time", "Street Name"]

# Filter out columns to keep
npvcs_df = npv_df[columns_to_keep]

# Display Data Frame Info
npvcs_df.info()
print()

# Display Data Frame
npvcs_df
```

### Step 3: Answer the Following Questions

#### 1. How many rows are in the data frame when it is read in the memory?

```python
# Determine the total number of rows of the Data Frame before pruning
num_rows = len(npvcs_df)

# Print the total number of rows before pruning the Data Frame
print(f"There are {num_rows:,} rows in the Data Frame")
```

#### 2. Remove rows with any missing data (i.e., a NaN value)

```python
# Determine the missing values per column in the Data Frame before pruning
missing_values = npvcs_df.isna().sum()

# Remove rows with any missing data in the Data Frame
npvcs1_df = npvcs_df.dropna()

# Determine the missing values per column in the Data Frame after pruning
missing_values1 = npvcs1_df.isna().sum()

# Print Info of missing values before pruning Data Frame
print("Missing values in each column before pruining:")
print(missing_values)
print()

# Print Info of missing values after pruning Data Frame
print("Missing values in each column after pruning:")
print(missing_values1)
print()

npvcs1_df
```

#### 3. How many rows remain after doing pruning?

```python
# Determine the total number of rows of pruned Data Frame
num_rows1 = len(npvcs1_df)

# Print the total number of rows before pruning the Data Frame
print(f"There are {num_rows1:,} rows in the Data Frame")
```

#### 4. If each parking ticket brings $100 into the city, missing data means that the ticket can be successfully contested, how much money might New York City lose as a result of missing data?

```python
# Calculate the number of rows with missing data and multiply by 100 to determine the potential loss in revenue
potential_loss = (len(npvcs_df) - len(npvcs1_df)) * 100

# Display the results
print(f"{len(npvcs_df) - len(npvcs1_df):,} rows in the Data Frame had missing values")
print()
print("New York City could potentially lose $ {:,.2f}".format(potential_loss), "in parking ticket fines")
```

#### 5. Let's instead assume that a ticket can only be dismissed if the license plate, state, car, make and/or street name are missing one or more of these. How many rows remain?

```python
# List of columns to check for missing values
columns_to_check = ["Plate ID", "Registration State", "Vehicle Make", "Street Name"]

# Remove rows where any of the specified columns have missing values
npvcs_df_cleaned = npvcs_df.dropna(subset=columns_to_check)

# Print the total number of rows remaining after pruning the Data Frame
total_rows_remaining = len(npvcs_df_cleaned)
print(f"There are {total_rows_remaining:,} rows remaining after pruning the Data Frame")
print()

# Display Info of pruned Data Frame
npvcs_df_cleaned.info()
print()

# Display pruned Data Frame
npvcs_df_cleaned
```

#### 6. Assuming $100/ticket, how much money would the city lose as a result of missing data?

```python
# Calculate the number of rows with missing data and multiply by 100 to determine the potential loss in revenue
potential_loss1 = (len(npvcs_df) -  len(npvcs_df_cleaned)) * 100

# Display the results
print(f"{len(npvcs_df) - len(npvcs_df_cleaned):,} rows in the Data Frame had missing values")
print()
print("New York City could potentially lose $ {:,.2f}".format(potential_loss1), "in parking ticket fines")
```
