# MORTGAGE CALCULATOR PROJECT

This program calculates the monthly payment made by a bank customer to pay off a mortgage loan and reports this amount as well as the total amount paid off at the end of the loan term.

## TIER 1

```python
from mortgage import Loan as ln

# Function to validate user input for numerical values
def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                raise ValueError
            return value
        except ValueError:
            print("Invalid input. Please enter a positive number.")

# Indicate Currency Symbol
currency_symbol = 'GHC'

# Indicate Percent Symbol
percent_symbol = '%'

# User inputs mortgage loan principal, annual interest rate, and years to pay off mortgage
mortgage_loan_principal = get_positive_float('Enter the mortgage loan principal: GHC')
percent_annual_interest = get_positive_float('Enter the percent annual interest:')
years_to_pay_off_mortgage = get_positive_float('Enter the years to pay off mortgage:')

# Create a Loan object with loan amount, interest rate, and loan term
loan = ln(principal=round(mortgage_loan_principal), interest=(percent_annual_interest/100), term=round(years_to_pay_off_mortgage))

# Calculate monthly loan payment
monthly_payment = loan.monthly_payment

# Calculate total payments
total_payment = monthly_payment * loan.term * 12

# Display the results
print("For a {}-year mortgage loan of {} {:.2f} at an annual interest of {:.2f} {}, you pay {} {:.2f} monthly and the total amount to be paid will be {} {:.2f}".format(
    round(years_to_pay_off_mortgage), currency_symbol, mortgage_loan_principal, percent_annual_interest, percent_symbol, currency_symbol, monthly_payment, currency_symbol, total_payment))
```

```python
# Function to validate user input for numerical values
def get_positive_float(prompt):
    while True:
        try:
            value = float(input(prompt))
            if value <= 0:
                raise ValueError
            return value
        except ValueError:
            print("Invalid input. Please enter a positive number.")

# Indicate Currency Symbol
currency_symbol = 'GHC'

# Indicate Percent Symbol
percent_symbol = '%'

# User inputs mortgage loan principal, annual interest rate, and years to pay off mortgage
mortgage_loan_principal = get_positive_float('Enter the mortgage loan principal: GHC')
percent_annual_interest = get_positive_float('Enter the percent annual interest: ')
years_to_pay_off_mortgage = get_positive_float('Enter the years to pay off mortgage: ')

# Calculate the monthly interest rate and months to pay off mortgage
percent_monthly_interest = percent_annual_interest / 12 / 100
months_to_pay_off_mortgage = years_to_pay_off_mortgage * 12

# Calculate the total monthly payment and total amount to be paid
total_monthly_payment = mortgage_loan_principal * (percent_monthly_interest * (1 + percent_monthly_interest) ** months_to_pay_off_mortgage) / ((1 + percent_monthly_interest) ** months_to_pay_off_mortgage - 1)
total_amount_paid = total_monthly_payment * months_to_pay_off_mortgage

# Display the results
print("For a {}-year mortgage loan of {} {:.2f} at an annual interest of {:.2f} {}, you pay {} {:.2f} monthly and the total amount to be paid will be {} {:.2f}".format(
    round(years_to_pay_off_mortgage), currency_symbol, mortgage_loan_principal, percent_annual_interest, percent_symbol, currency_symbol, total_monthly_payment, currency_symbol, total_amount_paid))
```
